# When should I use "groupby" in pandas? 

```python
import pandas as pd
```

```python
url = 'http://bit.ly/drinksbycountry'
drinks = pd.read_csv(url)
drinks.head()
```

```python
# The method .groupy() can help us to apply other methods result to specific groups.
# In this case we use .describe() across the columns
drinks.describe()
```

```python
# In this case we use .groupby() to make group based on the continent column and apply .describe() to them.
drinks.groupby('continent').describe()
```

```python
# If we'd only be interested in some columns and only in apply certain operations we could apply the method .agg()
# to compute a list of operations to an a particular column.
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```

#### To display plots of DataFrame using matplotlib we have to allow plots to appear in the notebook 

```python
%matplotlib inline
```

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean']).plot(kind='bar')
```
