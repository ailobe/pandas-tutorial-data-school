# How do I apply multiple filter criteria to a pandas DataFrame?

```python
import pandas as pd
```

```python
url = 'http://bit.ly/imdbratings'
movies = pd.read_csv(url)
movies.head()
```

```python
# We can add more than one condition by using the logical operators. These are: | for or, & for and, and ~ for not.
# The expressions must be grouped by using parentheses.
movies.loc[(movies.duration >= 200) & (movies.genre == 'Drama')] # & for and
```

```python
movies.loc[(movies.duration >= 200) | (movies.duration <= 90)] # | for or
```

```python
# remember we can sort the returned DataFramed using the sort_values() and passing an argument to sort by.
movies.loc[(movies.duration >= 200) | (movies.duration <= 90)].sort_values('duration') # sorted in ascending order.
```

```python
movies.loc[~ (movies.genre == 'Drama')] # ~ for not
```

```python
# The .isin() method returns each element in the Series that matches the values passed.
movies.loc[movies.genre.isin(['Crime','Adventure','Action'])]
```

```python
movies.DataFrame.isin('Adventure')
```
